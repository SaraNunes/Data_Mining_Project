```python
# Import MinMaxScaler from sklearn.preprocessing
from sklearn.preprocessing import MinMaxScaler
# # Defining the index of our dataset with the column Custid
group2.set_index(keys = group2['Custid'], inplace = True)
# Create a new dataset named group2_copy that will contain the columns MntPaint, MntSculpt, MntMusic, MntWriting, MntAudioV and MntLimited
group2_copy = group2[['MntPaint','MntSculpt','MntMusic','MntWriting','MntAudioV', 'MntLimited']]
# Create an instance of MinMaxScaler() named minmax_scaler.
minmax_scaler = MinMaxScaler()
#Use the .fit()method of minmax_scaler to fit the scaler to group2_copy and assign the results to the object scale
scale = minmax_scaler.fit(group2_copy)
#Use the .transform() method of scale to apply MinMaxScaler to group2_copy and assign it to the object group2_produsage.
group2_produsage = scale.transform(group2_copy)
#Create a new dataset named group2_produsage that will contain the following command : pd.DataFrame(group2_produsage, columns = group2_copy.columns). This will allow to convert the array created in the previous step to a DataFrame, and define the column names.
group2_produsage = pd.DataFrame(group2_produsage, columns = group2_copy.columns)
group2_produsage
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)

    # Fit model to samples
    model.fit(group2_produsage)

    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
```

```python
# When applying clustering, one important step we have to do is to normalize the data. For this we're gonna use the MinMaxScaler.
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(df)
df_norm = minmax_scaler.transform(df)
```

### Modelling
#### Apply SOM

```python
# The size of the map is 20 rows and 20 columns.
rows = 20
cols = 20
```

```python
som = MiniSom(
            x = rows, 
            y = cols, 
            input_len = len(data.columns), 
            learning_rate = 0.5,
            topology = "rectangular"  # "rectangular" or "hexagonal"   
            )
```

- learning_rate is the initial learning rate
- topology is  the topology of the map which is the outfit

```python
# The .pca_weights_init() method is used to initialize the weights to span the first two principal components.
som.pca_weights_init(df_norm)
```

```python
# Train the SOM with our data
som.train(data=df_norm, num_iteration=50)
```
